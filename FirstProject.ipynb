```python
shopping_results = results['shopping_results']
shopping_results
```

```python
lowest_price = shopping_results[0].get('price')[1:]
lowest_price_index = 0
```

```python
lowest_price = float(str(lowest_price).replace(',',''))
lowest_price
```

```python
number_of_results = int(len(shopping_results))
for i in range(number_of_results):
    print(f"Result {i+1}:")
    current_price  = shopping_results[i].get('price')
    current_price = float(str(current_price[1:]).replace(',',''))
    print(f"Title: {shopping_results[i].get('title', 'N/A')}")
    print(f"Price: {shopping_results[i].get('price', 'N/A')}")
    print(f"Source: {shopping_results[i].get('source', 'N/A')}")
    print(f'Link: {shopping_results[i].get("product_link","N/A")}')
    
    print("-" * 40)
    
    if (current_price < lowest_price):
        lowest_price = current_price
        lowest_price_index = i

```

```python
print('\n Best Deal:\n')           
print(f"Title: {shopping_results[lowest_price_index].get('title', 'N/A')}")
print(f"Price: {shopping_results[lowest_price_index].get('price', 'N/A')[1:]}")
print(f"Source: {shopping_results[lowest_price_index].get('source', 'N/A')}")
print(f'Link: {shopping_results[lowest_price_index].get("product_link","N/A")}')
print("-" * 40)    
```
